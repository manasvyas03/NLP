```python
from nltk.corpus import gutenberg

def analyze_gutenberg_texts():
    for fileid in gutenberg.fileids():
        num_chars = len(gutenberg.raw(fileid))
        num_words = len(gutenberg.words(fileid))
        num_sents = len(gutenberg.sents(fileid))
        avg_chars_per_word = num_chars / num_words
        ttr = len(set(gutenberg.words(fileid))) / num_words
        
        # Indent the print statements to be within the for loop
        print(f"Text: {fileid}")
        print(f"Total number of characters: {num_chars}")
        print(f"Total number of words: {num_words}")
        print(f"Total number of sentences: {num_sents}")
        print(f"Average number of characters per word: {avg_chars_per_word}")
        print(f"Type-token ratio (TTR): {ttr:.2f}")
        print()

analyze_gutenberg_texts()
```

```python
#List all the files available in the Gutenberg Corpus
gutenberg_files = gutenberg.fileids()
print("Gutenberg Corpus Files:",gutenberg_files)

#Load a specific text(eg. 'austen-emma.txt')
emma_text= gutenberg.raw('austen-emma.txt')
print("First 500 characters of 'Emma':\n", emma_text[:500])
```

```python
#tokenizing in gutenberg corpus 

from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import gutenberg

#tokenize words
emma_words = word_tokenize(emma_text)
print("first 20 words of 'Emma':\n")
print(emma_words[:20])

#tokenize sentences
emma_sentences = sent_tokenize(emma_text)
print("first 2 sentences of 'Emma':\n")
print(emma_sentences[:2])
```

```python
#frequency distribution

from nltk.probability import FreqDist
import matplotlib.pyplot as plt

fdist= FreqDist(emma_words)

print("Most Common Words:", fdist.most_common(10))

fdist.plot(30, title="Word Frequency Distribution in Emma")
```

```python
#Compare brown genres
from nltk.corpus import brown

def compare_brown_gernes():
    for category in brown.categories():
        num_words = len(brown.words(categories=category))
        num_sents = len(brown.sents(categories=category))
        freq_dist=FreqDist(brown.words(categories=category))
        
        print(f"Genre: {category}")
        print(f"Total number of words: {num_words}")
        print(f"Total number of sentences: {num_sents}")
        print(f"Most common words: {freq_dist.most_common(5)}")
        print()
compare_brown_gernes()
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

file_path = 'manas.html'

with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()

clean_text = re.sub(r'<.*?>', '', html_content)

tokens = word_tokenize(clean_text)

print(tokens)
```
